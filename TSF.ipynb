### Analysing profit based on states

```python
data['State'].unique()
```

```python
data["State"].value_counts()
```

```python
fig,ax=plt.subplots(figsize=(30,10))
sb.countplot(x=data['State'])
fig.autofmt_xdate()
```

```python
'''Calfornia state has the most sales, It means more people buy goods in california
for  office purpose as compared to other states.'''
#Wyoming state has the least sales, only one.
```

```python
data_state=data.groupby("State")[["Profit"]].mean()
data_state.head(10)
```

### Top 10 Highest profit states

```python
data_state=data_state.sort_values("Profit",ascending=False)
top_10=data_state.head(10)
top_10
```

```python
top_10['Profit'].plot(kind='bar',figsize=(15,5),color="blue")
plt.title("top 10 state with most profits")
```

```python
#vermont is the state which earns the highest average profit
```

### States with the lowest profit

```python
low_10=data_state.tail(10)
low_10
```

```python
low_10['Profit'].plot(kind='bar',figsize=(15,5),color="red")
plt.title("top 10 state with lowest profits")
```

```python
# Ohio state has the lowest averege profit.
```

# State wise analysis of sales

```python
data_sales=data.groupby("State")[["Sales","Discount"]].mean()
```

```python
data_sales.head()
```

```python
data_sales["Sales"].plot.pie(subplots=True,autopct="%1.2f%%",figsize=(50,20),labels=data_sales.index)
```

# top 10 highest selling states

```python
data_sales=data_sales.sort_values("Sales",ascending=False)
top_sales=data_sales.head(10)
top_sales["Sales"].plot(kind="bar",figsize=(15,10),color="green")
```

```python
# Wyoming state has the highest sales, beacause its total sale is only one.
```

```python
low_sales=data_sales.tail(10)
low_sales["Sales"].plot(kind="bar",figsize=(15,10),color="red")
```

```python
# South dakota has the lowest sales.
```

# analysing sales and discount simulatneously of top 10 profit states

```python
top_10=data.groupby("State")[["Profit","Sales","Discount"]].mean()
top_10=top_10.sort_values("Profit",ascending=False)
top_10=top_10.head(10)
top_10
```

```python
top_10.plot.pie(subplots=True,figsize=(20,20),autopct="%1.2f%%",label=top_10.index)
```

```python
''' vermont state has the highest profit, because it has no discount.... where as wyoming state
has the highest sales, but its profit is only 8.32%, meaning it gives the largest discount of 66.36%'''
```
